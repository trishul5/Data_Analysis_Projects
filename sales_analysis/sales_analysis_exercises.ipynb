```python
# Import necessary libraries
import pandas as pd
import glob
```

```python
# Merge all datasets into one complete CSV file
path = "/Users/trishulsathiyamoorthy/Desktop/sales_analysis/Sales_Data"
files = glob.glob(path + "/*.csv")

frames = []

for file in files:
    df = pd.read_csv(file, index_col=None, header=0)
    frames.append(df)

df = pd.concat(frames, axis=0, ignore_index=True)
df.columns
df
```

```python
df.to_csv("sales.csv", encoding="utf-8")
```

```python
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors="coerce")
```

```python
df
```

```python
# Find the month that had the most amount of sales
df["Month"] = df["Order Date"].dt.month
```

```python
# Convert columns into floats to be able to calculations
df["Price Each"] = pd.to_numeric(df["Price Each"], errors="coerce")
df["Quantity Ordered"] = pd.to_numeric(df["Quantity Ordered"], errors="coerce")
```

```python
# Create a column to the dataframe that calculates the cost
df["Total Cost"] = df["Price Each"] * df["Quantity Ordered"]
```

```python
df
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
best_month = df.groupby("Month")["Total Cost"].sum().keys()[11]
total_sales = df.groupby("Month")["Total Cost"].sum().max()
```

```python
print(f"Month {best_month} was the best month with ${'{0:.2f}'.format(total_sales)} dollars worth of sales")
```

```python
# EXERCISE: What US city had the highest number of sales
```

```python
def get_state(address):
    """Extracts the state from a single address"""
    
    if type(address) != str:
        return None
    else:
        try:
            return address.split(", ")[2].split(" ")[0]
        except IndexError:
            return None
```

```python
df["State"] = df["Purchase Address"].apply(get_state)
```

```python
def get_city(address):
    """Extracts the city from a single address"""
    
    if type(address) != str:
        return None
    else:
        try:
            return address.split(", ")[1]
        except IndexError:
            return None
```

```python
df["City"] = df["Purchase Address"].apply(get_city)
```

```python
df["Location"] = df["City"] + ", " + df["State"]
```

```python
total_sales = df.groupby("Location")["Total Cost"].sum().max()
city_with_maximum_sales = df.groupby("Location")["Total Cost"].sum().idxmax()
```

```python
print(f"{city_with_maximum_sales} had the most amount of sales which totaled ${'{0:.2f}'.format(total_sales)}")
```

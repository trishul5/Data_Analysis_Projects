```python
# Import necessary libraries
import pandas as pd
```

```python
# Read in dataset and assign it to a variable called courses
courses = pd.read_csv("/Users/trishulsathiyamoorthy/Desktop/data_analysis/udemy_analysis/udemy_dataset.csv")
courses.head()
```

# Find all the subjects in which Udemy offers courses in this dataset

```python
offered_courses = courses["subject"].unique()
for offered_course in offered_courses:
    print(offered_course, end="||")
```

# Find the subject with the most amount of courses

```python
subject_with_most_courses = courses.groupby("subject")["num_lectures"].sum().idxmax()
print(f"{subject_with_most_courses} is the subject with the most amount of courses")
```

# Show all the courses that are free of cost

```python
# Find type of is_paid column so I know I am working with a try boolean value
courses["is_paid"].dtype

free_courses = courses[courses["is_paid"] == False]
free_courses
```

# Show all the courses that are paid

```python
paid_courses = courses[courses["is_paid"] == True]
paid_courses
```

# Find the top 5 most selling courses

```python
# Approach: Sort the dataframe by the num_subscribers column and return back the head containing exactly 5 values
top_5_most_sold = courses.sort_values(by="num_subscribers", ascending=False).head(n=5)
top_5_most_sold
```

# Find the 5 least selling courses

```python
bottom_5_least = courses.sort_values(by="num_subscribers", ascending=True).head(n=5)
bottom_5_least
```

# Find the unique "Wind Speed" values in the dataset

```python
unique_values = len(weather_df["Wind Speed_km/h"].unique())

print(f"{unique_values} unique values in 'Wind Speed' column")
```

# Find the number of times when the Weather is exactly "Clear"

```python
clear_weather_occurrences = len(weather_df[weather_df["Weather"] == "Clear"])
print(f"Clear weather occurred a total of {clear_weather_occurrences} times")
```

# Find the number of times when the wind speed was exactly 4 km/h

```python
four_km_occurrences = len(weather_df[weather_df["Wind Speed_km/h"] == 4])
print(f"The wind speed was 4 km/h a total of {four_km_occurrences} times")
```

# Rename the column "Weather" of the dataframe to "Weather Condition"

# Find the mean visibility

```python
average_visibility = weather_df["Visibility_km"].mean()
print(f"The average visibility is {average_visibility}")
```

# Find the standard deviation of "Pressure" in the data

```python
pressure_standard_dev = weather_df["Press_kPa"].std()
print(f"The standard deviation of pressure is {pressure_standard_dev}")
```

# Find the variance of Relative Humidity

```python
relative_humidity_variance = weather_df["Rel Hum_%"].var()
print(f"The variance for relative humidity is {relative_humidity_variance}")
```

# Find all instances when "Snow" was recorded

```python
snow_instances = weather_df[weather_df["Weather Condition"].str.contains("Snow")]
snow_instances
print(f"There are a total of {len(snow_instances)} snow instances")
```

# Find all instances where "Wind Speed is above 24" and "Visibility is 25"

```python
above_24_and_equals_25 = weather_df[(weather_df["Wind Speed_km/h"] > 24) & (weather_df["Visibility_km"] == 25)]
above_24_and_equals_25
```

# Find the mean value of each column against the "Weather Condition"

```python
weather_df.groupby("Weather Condition").mean()
```

# Minimum and maximum values of each column grouped on Weather condition

```python
weather_df.groupby("Weather Condition").agg(["min", "max"])
```

# Records where weather condition is fog

```python
fog_instances = weather_df[weather_df["Weather Condition"].str.contains("Fog")]
fog_instances
```

# Instances where weather is clear or visibility is above 40

```python
dual_condition = weather_df[ (weather_df["Weather Condition"].str.contains("Clear")) | (weather_df["Visibility_km"] > 40) ]
dual_condition.tail(50)
```

# Instances of clear weather and relative humidity is greater than 50 OR visibility is above 40

```python
filt = ( ((weather_df["Weather Condition"].str.contains("Clear")) & (weather_df["Rel Hum_%"] > 50)) | (weather_df["Visibility_km"] > 40))

output_df = weather_df[(filt)]
output_df
```
